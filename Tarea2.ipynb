### c) Generar distribución Gaussiana
Calcule el promedio y la desviación estándar de ambas columnas (**pmra** y **pmdec**) y guárdelos en las variables que se indican.

```python
# pmra, promedio y desviación estándar
mu_pmra = pmra.mean()
std_pmra = pmra.std()

# pmdec, promedio y desviación estándar
mu_pmdec = pmdec.mean()
std_pmdec = pmdec.std()

print(mu_pmra)
print(mu_pmdec)
```

Con los promedios y desviaciones estandar obtenidas, simule una distribución gaussiana para cada eje del movimieto propio (pmra y pmdec). Recuerda que la distribución gaussiana es de la forma:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$


Por lo que necesitamos los valores de $x, \mu$ y $\sigma$ para cada caso.

**En ascención recta**:
1. Crea un arreglo de 300 números que estén entre -1.5 y 1. Este sería el $x$
2. Con $\mu$ y $\sigma$ correspondientes a pmra calculados en la celda de arriba, genere los valores $f(x)$. Para eso, utilize la función **norm** de **scipy.stats** que está importada en la siguiente celda. Usa el método **pdf** (probability density function)



**En declinación:**
Haga lo mismo, pero el $x$ sería 300 números que estén entre -2.5 y 0.5.

```python
from scipy.stats import norm

# pmra (x, f(x))
x_pmra = np.linspace(-1.5, 1, num=300)
fx_pmra = norm.pdf(x_pmra,loc=mu_pmra,scale=std_pmra)

# pmdec (x, f(x))
x_pmdec = np.linspace(-2.5, 0.5, num=300)
fx_pmdec = norm.pdf(x_pmdec,loc=mu_pmdec,scale=std_pmdec)
```

### d) Graficar histogramas de la distribución de movimientos propios junto a la distribución Gaussiana

Grafique con matplotlib un hitograma (busque qué método de matplotlib sirve) de pmra y sobre él grafique la gaussiana correspondiente.

```python
# gráfico de distribución de pmra
plt.title("Distribución de PMRA")
pmra.plot.hist(xlabel="PMRA [mas/yr]",ylabel="Frecuencia",bins=50,label="distribución de pmra")     # histograma
plt.plot(x_pmra,fx_pmra,label="distribución normal")      # distribución normal
plt.legend()
```

Notará que la gaussiana apenas se ve en el gráfico, esto es porque la gaussiana está normalizada (el área bajo la curva es 1), pero esto no es así para el histograma. El problema se puede arreglar incluyendo el parámetro **density=True** en el método que grafica el histograma.

Repita la figura anterior y ahora agréguele a cada gráfica (pmra y pmdec) una franja vertical que abarque $\mu \pm \sigma$ utilizando **plt.axvspan()**

```python
# gráfico de distribución de pmra
plt.figure()
plt.title("Distribución de PMRA")
plt.axvspan(mu_pmra-std_pmra,mu_pmra+std_pmra,alpha=0.3,label="mu +- sigma (pmra)")      # franja vertical, mu +- std
pmra.plot.hist(xlabel="PMRA [mas/yr]",ylabel="Frecuencia",bins=50,label="distribución de pmra",density=True)      # histograma
plt.plot(x_pmra,fx_pmra,label="distribución normal")      # distribución normal
plt.legend()

# gráfico de distribución de pmdec
plt.figure()
plt.title("Distribución de PMDEC")
plt.axvspan(mu_pmdec-std_pmdec,mu_pmdec+std_pmdec,alpha=0.3,label="mu +- sigma (pmdec)")  # franja verical, mu +- std
pmdec.plot.hist(xlabel="PMDEC [mas/yr]",ylabel="Frecuencia",bins=50,label="distribución de pmdec",density=True)    # histograma
plt.plot(x_pmdec,fx_pmdec,label="distribución normal")    # distribución normal
plt.legend()
```

### b) Nivel de ruido:

En esta parte estimaremos el nivel de ruido de la la observación 1 de una forma distinta. Para esto siga los sigueintes pasos:

1. Tome la columna **obs_1** y guarde en la variable **ruido** un subconjunto de datos centrados en la media de **obs_1** con un margen de $\pm 3 \sigma$, con $\sigma$ la desviación estándar de **obs_1**.

Note que este subconjunto corresponde a ruido principalmente.

```python
# desviación estándar
std_obs_1 = data["obs_1"].std()

# crear lista ruido
ruido = []
for i in obs_1:   # iteración para analizar cada dato de obs_1
  if (i >= mu_obs_1 - 3*std_obs_1) and (i <= mu_obs_1 + 3*std_obs_1):   # condición para pertenecer al subconjunto ruido
    ruido.append(i)   # guardar el elemento en la lista

# lista -> arreglo
ruido = np.array(ruido)
```

2. Reste la media del ruido al espectro para centrar y grafique el espectro resultante.

```python
# promedio ruido
mu_ruido = ruido.mean()

# gráfico del espectro
plt.figure(figsize=[10,5])
plt.grid(True)
plt.title("Espectro centrado en potencia=0")
plt.xlabel("Longitud de onda")
plt.ylabel("Potencia")
plt.plot(Wlth,obs_1-mu_ruido,label="espectro")
plt.legend()
```

3. Calcule $\sigma$ del ruido y grafique un margen de $\pm n*\sigma$ (con $n$ entero) sobre el espectro como dos líneas horizontales. ¿Qué valor de $n$ le parece adecuado y por qué? este será el nivel de ruido.

**Respuesta:**

Ingresé n=2 como valor adecuado, pues parece cubrir lo justo y necesario del ruido, lo cubre con un pequeño margen extra, para así no pasar a llevar las que proablemente sean líneas de emisión (parecen haber 3)

```python
n=2   # valor de n
std_ruido = ruido.std()   # desviación estándar del ruido


# gráfico del espectro
plt.figure(figsize=[10,5])
plt.grid(True)
plt.title("Espectro centrado en potencia=0 (n=2)")
plt.xlabel("Longitud de onda")
plt.ylabel("Potencia")
plt.plot(Wlth,obs_1-mu_ruido,label="espectro")

# líneas horizontales
ymin = n*std_ruido    # valor mínimo vertical
ymax = -n*std_ruido   # valor máximo vertical
xmin = min(Wlth)      # valor mínimo horizontal
xmax = max(Wlth)      # valor máximo horizontal

# plotear líneas horizontales
plt.hlines(ymin,xmin,xmax,colors="r",label="margen + n*sigma_ruido")
plt.hlines(ymax,xmin,xmax,colors="r",label="margen - n*sigma_ruido")
plt.legend()
```
